```python
df_puebla = pd.read_csv('./data/puebla.csv')
df_global = pd.read_csv('./data/global.csv')
```

```python
df_puebla= df_puebla.set_index(['year']).rename(columns={'avg_temp':'puebla_avg_temp'})
```

```python
df_global= df_global.set_index(['year']).rename(columns={'avg_temp':'global_avg_temp'})
```

```python
df = df_puebla.merge(df_global, left_index=True, right_index=True, how = 'right')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def plot_mvg(roller=5):
    df_mavg = df.rolling(roller).apply(np.nanmean).rename(columns={'puebla_avg_temp':'puebla_20y_mavg_temp',
                                                              'global_avg_temp':'global_20y_mavg_temp'})
    df_mavg.plot(style='.-').set_ylabel("temperature")
    
    return df_mavg
```

```python
df_mavg = plot_mvg(10)
```

```python
def plot_mvg_ratio(roller=5):
    fig=plt.figure(figsize=(10,8))
    plt.rc('xtick', labelsize=14) 
    plt.rc('ytick', labelsize=14) 
    df_aux = (df['puebla_avg_temp']/df['global_avg_temp']).rolling(roller).apply(np.nanmean)
    g = df_aux.plot(style='.-')
    g.set_ylabel("temperature ratio", fontsize=16 )
    fig.savefig('./plots_tables/puebla_vs_global.png')
plot_mvg_ratio(50)
```
